```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_jobs.csv')
df.head(5)
```

```python
df.info()
```

```python
type(df['job_skills'][1])
```

### Data cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.head(3)
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
df.head(3)
```

```python
df_DA = df[df['job_title_short']=='Data Analyst'].copy()
df_DA.head(2)
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
df_DA.head(3)
```

```python
df_DA_explode = df_DA.explode('job_skills')
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(
    index='job_posted_month_no',
    columns='job_skills',
    aggfunc='size'
).fillna(0)

df_DA_pivot
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index] 
df_DA_pivot.drop('Total',inplace=True)
df_DA_pivot
```

```python
df_DA_pivot.iloc[:,:5].plot(kind='line')
plt.title('Top 5 skills for Data Analysts per month')
plt.ylabel('Count')
plt.xlabel('')
```

```python
import ast
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('data_jobs.csv')
df =dataset.copy()
df.head(3)
```

### Data cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df.head(3)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%')
plt.title('Work from home status')
plt.ylabel('')
```

```python
df['job_title_short'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Job titles')
plt.ylabel('')
```

```python
df['job_title_short'].value_counts().plot(kind='barh')
plt.title('Job titles')
plt.gca().invert_yaxis()
plt.ylabel('')
```

### Pie chart

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home':'Work from home',
    'job_no_degree_mention':'Job Degree Req.',
    'job_health_insurance':'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(),autopct='%1.1f%%',startangle=90,labels=['False','True'])
    ax[i].set_title(title)
    
```

```python
df_exploded = df.explode('job_skills')
df_exploded
```

```python
skill_stats = df_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills','count'),
    median_salary = ('salary_year_avg','median')
)
skill_stats = skill_stats.sort_values(by='skill_count',ascending=False).head(10)
skill_stats
```

```python
skill_stats.plot(kind='scatter',x='skill_count',y='median_salary')
plt.xlabel('Count of job position')
plt.ylabel('Median yearly salary')
plt.title('Salary vs. Count of job postings for top 10 skills')
plt.tight_layout()

for i, txt in enumerate(skill_stats.index):
    plt.text(skill_stats['skill_count'].iloc[i], skill_stats['median_salary'].iloc[i],txt)
```
