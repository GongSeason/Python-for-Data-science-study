```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name':['Anna','Karen','John','Alice','Kevin','Sanna','Bob','Emily'],
    'Age':[35,30,57,65,25,19,20,65],
    'Salary':[20000,60000,145000,170000,300000,10000,220000,120000],
    'Department':['Tech','Tech','Tech','Healthcare','Operation','Operation','Tech','Tech']
})
df
```

### Sorting your data


```python
# ascending: smallest values at the top and values increasing
df_ascending = df.sort_values(by='Salary',ascending=True)
df_ascending
```

```python
# descending: largest values at the top and values decreasing
df_descending = df.sort_values(by='Salary',ascending=False)
df_descending
```
